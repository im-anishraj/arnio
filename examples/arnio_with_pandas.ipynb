# Arnio + Pandas Interoperability
This notebook demonstrates how to:
- Start with messy data in pandas
- Clean and normalize it using Arnio's pipeline
- Convert it back to pandas for analysis

```python
!pip install arnio pandas
```

## Step 1: Create Messy Dataset (pandas)

```python
import pandas as pd

import arnio as ar

df = pd.DataFrame({
    'name': [' Alice ', 'Bob', 'CHARLIE', None],
    'age': ['25', '30', None, '40'],
})
print('Original Data:')
print(df)
```

## Step 2: Convert pandas → Arnio

```python
frame = ar.from_pandas(df)
```

## Step 3: Clean Data using Arnio Pipeline

```python
cleaned = ar.pipeline(
    frame,
    [
        ('drop_nulls',),
        ('strip_whitespace',),
        ('normalize_case', {'case_type': 'lower'}),
        ('cast_types', {'mapping': {'age': 'int64'}}),
    ],
)
```

## Step 4: Convert Arnio → pandas

```python
clean_df = ar.to_pandas(cleaned)
print('Cleaned Data:')
print(clean_df)
```

## Step 5: Use pandas for Analysis

```python
print('Summary Statistics (pandas):')
print(clean_df.describe())
```
